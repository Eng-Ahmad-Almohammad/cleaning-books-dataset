## All Year of Publication column data should be integers and we should specify all years that are satisfy our study as example in this dataset the acceptable yars for me are in range 0 < years < 2025

## What we will do with rows that have number of years out of our range we have many choices:
1- We can drop them.

2- We can replace the year value with the most frequent year.

3- We can calculate the mean of the years and assign it to that rows // I chose this approach

## If you noticed that the range of the ages is 0 <= Age <= 244 which is not acceptable for me the acceptable range is 5 <= Age <= 244.
## As we did with the Years of Publication we have 3 approaches and I chosed to replace the out of range data with the mean of ages.

```python
df_users.loc[(df_users['Age'] > 90) |( df_users['Age'] < 5) , 'Age'] = np.NaN
```

```python
df_users.loc[df_users['Age'] > 89]
```

```python
user_mean_age = df_users['Age'].mean()
user_mean_age
```

```python
user_mean_age = round(user_mean_age)
user_mean_age
```

```python
df_users['Age'].fillna(user_mean_age , inplace = True)
```
